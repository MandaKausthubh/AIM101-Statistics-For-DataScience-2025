# Notebook 2: Intermediate Matplotlib Concepts


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 8, 5, 10]
data = np.random.randn(1000)
```

### Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0, 0].plot(x, y, label='Line')
axes[0, 0].set_title("Line Plot")
axes[0, 0].legend()

axes[0, 1].scatter(x, y, color='green')
axes[0, 1].set_title("Scatter Plot")

axes[1, 0].bar(categories, values, color='orange')
axes[1, 0].set_title("Bar Chart")

axes[1, 1].hist(data, bins=15, color='purple')
axes[1, 1].set_title("Histogram")
plt.tight_layout()
plt.show()
```

### Adding gridlines, customizing ticks, and colors

```python
plt.plot(x, y, linestyle='--', color='magenta', marker='o')
plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.show()
```

# Notebook 3: Dataset Analysis with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load dataset (example: Iris dataset)

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

### Scatter plot for sepal length vs sepal width


```python
plt.scatter(iris['sepal_length'], iris['sepal_width'], c='blue', alpha=0.5)
plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

### Bar plot for average petal length per species


```python
species_means = iris.groupby('species')['petal_length'].mean()
plt.bar(species_means.index, species_means)
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length")
plt.show()
```

### Histogram for petal width


```python
plt.hist(iris['petal_width'], bins=10, color='green')
plt.title("Distribution of Petal Width")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.show()
```

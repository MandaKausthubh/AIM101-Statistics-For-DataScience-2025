# Central Limit Theorem
### What is the Central Limit Theorem (CLT)?
The Central Limit Theorem states that when we take many samples from a population, the distribution of the sample means will approach a normal distribution (bell-shaped curve) as the sample size increases, regardless of the population's distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### What Are We Doing?
1. We create a large population dataset.
2. We take multiple samples from this population.
3. We calculate the means of these samples and plot their distribution.

### Full dataset

```python
population = np.random.normal(50, 10, size=10000)
```

### Sample means from multiple samples

```python
num_samples = 1000
sample_size = 50
sample_means = [np.mean(np.random.choice(population, size=sample_size, replace=False)) for _ in range(num_samples)]
```

### Plot histogram of sample means
### Why is This Important?
The Central Limit Theorem is fundamental in statistics because it allows us to make predictions about population parameters using sample data. It explains why many statistical methods work, even with non-normally distributed data.

```python
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=30, color='purple', alpha=0.7, density=True)
plt.title("Central Limit Theorem: Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid()
plt.show()
```

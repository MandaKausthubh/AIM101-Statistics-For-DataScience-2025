# Frequency Distribution in Full Dataset vs. Sample

```python
import numpy as np
import matplotlib.pyplot as plt
```

### What Are We Doing?
1. We create a large dataset and plot its frequency distribution.
2. We take random samples from this dataset and plot their frequency distributions.
3. We compare how the sample distributions change as we increase the sample size.

### Full dataset

```python
large_data = np.random.normal(50, 10, size=10000)
```

### Random samples of varying sizes
### Why is This Important?
This exercise helps us see how smaller samples can represent the larger dataset. As the sample size increases, the sample distribution tends to resemble the original dataset, showing the power of sampling.


```python
sample_sizes = [50, 200, 1000, 2000, 5000, 10000]
plt.figure(figsize=(15, 10))

for i, size in enumerate(sample_sizes, 1):
    sample = np.random.choice(large_data, size=size, replace=False)

    # Create a 2x3 grid of subplots
    plt.subplot(2, 3, i)

    # Plot histogram for the sample
    plt.hist(sample, bins=100, alpha=0.7, color='red', density=True, label="Sample")

    # Plot histogram for the full dataset
    plt.hist(large_data, bins=100, alpha=0.3, color='blue', density=True, label="Full Dataset")

    plt.title(f"Sample Size: {size}")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()

plt.tight_layout()
plt.show()
```

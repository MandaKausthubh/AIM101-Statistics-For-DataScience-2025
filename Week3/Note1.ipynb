# Median and Cumulative Absolute Distance Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

### What Are We Doing?
1. We generate a random dataset.
2. We calculate cumulative absolute distances for different candidate centers.
3. We plot these distances to see where the minimum occurs

### Generate random data

```python
data = np.random.randint(1, 100, size=50)
```

### Candidate centers to evaluate

```python
candidate_centers = np.linspace(min(data), max(data), 200)
# we are calculating the sum of the squared distances of each data point to the candidate center
```

### What is Cumulative Absolute Distance?
Cumulative absolute distance measures how far each point in the dataset is from a candidate center. We compute the total distance for each center and find which center has the smallest total distance.


```python
cumulative_distances = [np.sum(np.abs(data - c)) for c in candidate_centers]
# we are calculating the sum of the absolute distances of each data point to the candidate center
```

## Why is This Plot Important?
This plot helps us visualize how the median is the best center point for minimizing the total distance of all points in the dataset. It shows that the median is more robust to outliers than the mean.


```python
plt.figure(figsize=(8, 5))
plt.plot(candidate_centers, cumulative_distances, label="Cumulative Absolute Distance")
plt.axvline(np.median(data), color='red', linestyle='--', label="Median")
plt.xlabel("Center")
plt.ylabel("Cumulative Absolute Distance")
plt.title("Cumulative Absolute Distance vs Center")
plt.legend()
plt.show()
```

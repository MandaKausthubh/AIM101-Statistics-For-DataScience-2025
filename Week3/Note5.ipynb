# Law of Large Numbers
### What is the Law of Large Numbers?
The Law of Large Numbers states that as we take larger samples from a population, the sample mean (average) will get closer to the population mean.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### What Are We Doing?
1. We create a large population dataset.
2. We take samples of increasing sizes.
3. We calculate the mean of each sample and plot it against the sample size.

### Full dataset mean

```python
population = np.random.normal(50, 10, size=10000)
pop_mean = np.mean(population)
```

### Sample means for increasing sample sizes

```python
#generates an array of evenly spaced values within a specified range, START, STOP, STEP
sample_sizes = np.arange(10, 500, 10)
sample_means = []
```

```python
for s in sample_sizes:
    # Generate a number of samples and calculate their means
    samples = [np.mean(np.random.choice(population, size=s, replace=False)) for _ in range(1000)]
    sample_means.append(np.mean(samples))  # Store the mean of the sample means
```

### Why is This Plot Important?
This concept is crucial because it shows that larger samples provide more reliable estimates of the population characteristics. It helps us understand why big data is important in statistics.

```python
plt.figure(figsize=(8, 5))
plt.plot(sample_sizes, sample_means, label="Sample Mean", marker='o')
plt.axhline(pop_mean, color='red', linestyle='--', label="Population Mean")
plt.xlabel("Sample Size")
plt.ylabel("Mean")
plt.title("Law of Large Numbers")
plt.legend()
plt.grid()
plt.show()
```

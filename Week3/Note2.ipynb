# Relationship Between Mean and Median in Skewed Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm
```

## Generate skewed datasets

### What is Skewness?
Skewness refers to how much a distribution leans to one side. If a distribution has:
- A long tail on the left, it's called **left-skewed**.
- A long tail on the right, it's called **right-skewed**.

```python
def generate_skewed_dataLib(a, size=500):  # 'a' controls the skewness
    return skewnorm.rvs(a, size=size)

def generate_skewed_dataNoLib(a, size=500):
    # Generate a normal distribution
    data = np.random.normal(0, 1, size)
    data = np.abs(data) * a
    return data
```

### What Are We Doing?
1. We create two skewed datasetsâ€”one that leans left and another that leans right.
2. We plot histograms for both datasets.
3. We add lines for the mean and median.

## Skewed left (negative skew)

```python
data_left1 = generate_skewed_dataLib(-5)
data_left2 = generate_skewed_dataNoLib(-5)
```

## Skewed right (positive skew)

```python
data_right1 = generate_skewed_dataLib(5)
data_right2 = generate_skewed_dataNoLib(5)
```

## Plot histograms with mean and median

```python
data_sets = [
    (data_left1, "Skewed Left With Library"),
    (data_left2, "Skewed Left Without Library"),
    (data_right1, "Skewed Right With Library"),
    (data_right2, "Skewed Right Without Library")
]
```

## Create a 2x2 grid of subplots
### Why is This Important?
This helps us understand how the mean and median behave in different situations. In skewed distributions, the mean can be pulled towards the tail, while the median remains a more stable center.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

for ax, (data, title) in zip(axs.flatten(), data_sets):
    ax.hist(data, bins=30, alpha=0.7, label="Data")
    ax.axvline(np.mean(data), color='blue', linestyle='--', label="Mean")
    ax.axvline(np.median(data), color='red', linestyle='--', label="Median")
    ax.set_title(title)
    ax.legend()

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

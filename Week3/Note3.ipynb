# Histograms and Frequency Plots with Different Bin Sizes

```python
import numpy as np
import matplotlib.pyplot as plt
```

### What Are We Doing?
1. We generate a random dataset.
2. We create histograms with different bin sizes.
3. We observe how changing the bin size affects the shape of the histogram.

### Generate random data

```python
data = np.random.normal(loc=50, scale=10, size=1000)
```

### Plot histograms with varying bin sizes
Bins are the intervals we use to group the data. The width of these bins affects how we see the data.

```python
bin_sizes = [10, 20, 50]
```

### Why is This Important?
By experimenting with bin sizes, we learn how to better visualize and interpret data. This helps us understand the underlying patterns and distributions.

```python
plt.figure(figsize=(15, 5))
for i, bins in enumerate(bin_sizes, 1):
    plt.subplot(1, len(bin_sizes), i)
    plt.hist(data, bins=bins, alpha=0.7, color='skyblue')
    plt.title(f"Histogram with {bins} bins")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

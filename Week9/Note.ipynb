```python
p_vals = np.linspace(0.01, 0.99, 100)
n_vals = [10, 20, 50, 100, 500, 1000]
rows, cols = 2, 3
```

```python
# Create figure and axes
fig, axes = plt.subplots(rows, cols, figsize=(18, 10))
axes = axes.flatten()

# Plot each sample size's MSE curves
for idx, n in enumerate(n_vals):
    ax = axes[idx]

    # MSE of MLE estimator: p = m/n
    mse_mle = (p_vals * (1 - p_vals)) / n

    # MSE of smoothed estimator: p = (m+1)/(n+2)
    mse_smoothed = (n * p_vals * (1 - p_vals) + (1 - 2 * p_vals)**2) / (n + 2)**2

    # Plotting both MSEs
    ax.plot(p_vals, mse_mle, label=r"$\mathrm{MSE}(\hat{p})$", color='blue')
    ax.plot(p_vals, mse_smoothed, label=r"$\mathrm{MSE}(\tilde{p})$", color='red', linestyle='--')

    ax.set_title(f"Sample Size: n = {n}", fontsize=12)
    ax.set_xlabel("True Probability $p$")
    ax.set_ylabel("Mean Squared Error")
    ax.grid(True)
    ax.legend()
# Remove unused subplot (if layout has extra axes)
for i in range(len(n_vals), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```
